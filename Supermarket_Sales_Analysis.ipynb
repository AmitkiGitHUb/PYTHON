**Creating dummies for Gender column for getting numeric data for seaborn bar plot**

**Plotting heatmap to check if any correlation between columns**

```python
sns.heatmap(df.corr())
```

```python
sns.heatmap(df.select_dtypes(include=np.number).corr())
```

```python
# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=np.number)

# Calculate the correlation matrix
corr_matrix = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))  # Adjust figure size as needed
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
# Select numerical columns, including 'Female' and 'Male'
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
# Add 'Female' and 'Male' if not already present
for col in ['Female', 'Male']:
    if col not in numerical_cols:
        numerical_cols.append(col)

numerical_df = df[numerical_cols]

# Calculate the correlation matrix
corr_matrix = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(14, 10))  # Adjust figure size as needed
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features including Gender')
plt.show()
```
